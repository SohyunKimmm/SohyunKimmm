```python
#독립성 검정
#귀무가설(Ho)= 두 집단(변수)이 독립적이다. 
#대립가설(H1)= 두 집단(변수)이 독립적이지 않다. 
#p-value<0.05 : 대립가설 채택; 두 변수가 독립적이지 않다 = '독립성이 깨졌다', 두 변수 모두 반영하면 X


#등분산성 검정
#귀무가설(H0): 두 표본이 등분산을 가지고 있다. (등분산이다)
#대립가설(H1): 두 표본이 등분산이 아니다.
#p-value<0.05: 대립가설 채택; 두 표본이 등분산이 아니다 = '등분산성이 깨졌다' 
#등분산이 깨졌으면, t test(평균을 비교) 혹은 ANOVA테스트에서 정확한 비교는 어려움
```

```python
#ANOVA test
#세 집단 이상 '평균'비교라는 걸 잊지말기!! ('분산'비교 X)

import pingouin as pg
import scikit_posthocs
import scipy as sp
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
df2["구매유형"].describe()
```

```python
df2.groupby("구매유형").mean()
```

```python
df1 = df2[["구매유형", "총_매출액"]]
pd.options.display.float_format = '{:.3f}'.format #결과값 소숫점을 3째자리까지 표시

type1 = []
for i in range(1,5,1):  #4개 집단의 평균 비교?(인덱스 #1,2,3,4)
    type1.append(df1[df1.구매유형==i].총_매출액) #구매유형별(i=1,2,3,4)로 '총_매출액'을 type1 리스트 안에 더해감
sp.stats.levene(type1[0], type1[1], type1[2], type1[3])  #for문이 돌면서 인덱스1,2,3,4가 0,1,2,3으로 바뀜

#pvalue가 0.05보다 작기 때문에 대립가설 채택(등분산이 아니다.)
```
