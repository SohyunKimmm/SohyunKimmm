```python
article_txt = open("/content/drive/MyDrive/뉴진스그램comment.txt", "rt", encoding = "UTF-8").read()
```

```python
article_txt
```

```python
from konlpy.tag import Okt
from collections import Counter
```

```python
okt = Okt()
line = []

line = okt.pos(article_txt) #article text안에 품사를 붙여주는 함수
n_adj = []
for word, tag in line:
    if tag in ["Noun", "Adjective"]: #명사, 형용사만 뽑음
        n_adj.append(word)
print(n_adj)
```

```python
stop_words = "좀 거 임 왜 팀 더 임 때 저 "
stop_words = set(stop_words.split(" "))
n_adj = [word for word in n_adj if not word in stop_words]
```

```python
counts = Counter(n_adj)
tags = counts.most_common(50)
print(counts)
print(tags)
```

```python
from wordcloud import WordCloud
```

```python
font = "/content/drive/MyDrive/malgun.ttf"
word_cloud = WordCloud(font_path = font, background_color = "white", max_font_size = 100, 
                       colormap = "PuBuGn").generate_from_frequencies(dict(tags))
import matplotlib.pyplot as plt
plt.figure(figsize = (15,12))
plt.imshow(word_cloud)
plt.axis("off")
```
