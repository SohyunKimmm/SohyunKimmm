```python
import pandas as pd
import pingouin as pg
import scikit_posthocs
import scipy as sp
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
sp_sales = pd.read_csv("/content/drive/MyDrive/230103_Ashopping2.csv", encoding = "cp949")
sp_sales
```

```python
sp_sales.columns
```

```python
sp_sales["고객_나이대"].describe()
#min=1, max=9 
```

```python
#구매_카테고리_수 별 1회_평균매출액 평균
sp_sales.groupby("고객_나이대").mean()["매장_만족도"]
```

```python
#등분산test(1)
#귀무가설: 등분산이다.
#대립가설: 등분산이 아니다.

sp_sales_anova = sp_sales[["고객_나이대", "매장_만족도"]]
pd.options.display.float_format = '{:.4f}'.format #결과값 소숫점을 4째자리까지

type1 = []
for i in range(1,10,1): #9개의 집단 평균비교(#1,2,3,4,5,6,7,8,9)
    type1.append(sp_sales_anova[sp_sales_anova.고객_나이대==i].매장_만족도)
sp.stats.levene(type1[0],type1[1],type1[2],type1[3],type1[4],type1[5],type1[6],type1[7],type1[8])

#결과값
#pvalue = 0.395 (>0.05)
#귀무가설 채택 -> '등분산이다'
```

```python
#등분산test(2)
pg.homoscedasticity(dv= "매장_만족도", group="고객_나이대", data = sp_sales_anova)

#결과값
#pvalue = 0.3959 (>0.05)
#귀무가설 채택 -> '등분산이다'
```

```python
#등분산인 경우(등분산 가정)
#귀무가설: 고객 나이대별 매장만족도의 평균이 같다. 
#대립가설: 고객 나이대별 매장만족도의 평균이 같지 않다. 

pg.anova(dv= "매장_만족도", between="고객_나이대", data = sp_sales_anova, detailed=True)

#결과값
#pvalue = 0.45 (>0.05)
#귀무가설 채택 -> 고객 나이대별 매장만족도의 평균이 같다. 
```
