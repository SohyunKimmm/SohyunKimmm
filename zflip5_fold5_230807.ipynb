```python
import pandas as pd

# txt 파일의 경로 설정 (아래의 '파일경로.txt'는 실제 파일 경로로 변경해야 합니다)
file_path = '/content/z플립_폴드_5.txt'

# txt 파일 읽기
with open(file_path, 'r') as file:
    article_txt = file.read()

# 데이터 확인
print(article_txt)
```

```python
from konlpy.tag import Okt
from collections import Counter
```

```python
okt = Okt()
line = []

line = okt.pos(article_txt)
n_adj = []
for word, tag in line:
  if tag in ["Noun", "Adjective"]:
    n_adj.append(word)
print(n_adj)
```

```python
stop_words = ""
stop_words = set(stop_words.split(" "))
n_adj = [word for word in n_adj if not word in stop_words]
```

```python
counts = Counter(n_adj)
tags = counts.most_common(50)
print(counts)
print(tags)
```

```python
from wordcloud import WordCloud
```

```python
font = "/content/drive/MyDrive/malgun.ttf"
word_cloud = WordCloud(font_path = font, background_color = "black", max_font_size = 300,
                       colormap = "Blues").generate_from_frequencies(dict(tags))
import matplotlib.pyplot as plt
plt.figure(figsize = (10,8))
plt.imshow(word_cloud)
plt.axis("off")
```

```python
import pandas
from googleapiclient.discovery import build
```

##1. 영상(1)_"너무 예뻐서 무조건 살 겁니다. 갤럭시 Z 플립5&폴드5 실물로 본 느낌은?"

```python
comments = list()
api_obj = build('youtube', 'v3', developerKey="AIzaSyB77UMcjZkgg5MnK7LnyC25yC_PVGM567E")
response = api_obj.commentThreads().list(part='snippet,replies', videoId="andZAenBIyI", maxResults=100).execute()
```

```python
while response:
    for item in response['items']:
        comment = item['snippet']['topLevelComment']['snippet']
        comments.append([comment['textDisplay'], comment['authorDisplayName'], comment['publishedAt'], comment['likeCount']])

        if item['snippet']['totalReplyCount'] > 0:
            for reply_item in item['replies']['comments']:
                reply = reply_item['snippet']
                comments.append([reply['textDisplay'], reply['authorDisplayName'], reply['publishedAt'], reply['likeCount']])

    if 'nextPageToken' in response:
        response = api_obj.commentThreads().list(part='snippet,replies', videoId="andZAenBIyI", pageToken=response['nextPageToken'], maxResults=100).execute()
    else:
        break
```

```python
df = pandas.DataFrame(comments)
df.to_excel('z5_results.xlsx', header = ['comment', 'author', 'date', 'num_likes'], index = None)
```
