##유튜브 댓글 크롤링

```python
!pip install google-api-python-client
```

```python
import pandas
from googleapiclient.discovery import build
```

```python
comments = list()
api_obj = build('youtube', 'v3', developerKey= "AIzaSyDwp5MaZ_EfPQC8AIZENIp2IWxX65iBV8o")
response = api_obj.commentThreads().list(part='snippet,replies', videoId="NxZDcqhrdAc", 
                                         maxResults=2500).execute()
```

```python
while response:
    for item in response['items']:
        comment = item['snippet']['topLevelComment']['snippet']
        comments.append([comment['textDisplay'], comment['authorDisplayName'], 
                         comment['publishedAt'], comment['likeCount']])
 
        if item['snippet']['totalReplyCount'] > 0:
            for reply_item in item['replies']['comments']:
                reply = reply_item['snippet']
                comments.append([reply['textDisplay'], reply['authorDisplayName'], 
                                 reply['publishedAt'], reply['likeCount']])
 
    if 'nextPageToken' in response:
        response = api_obj.commentThreads().list(part='snippet,replies', 
                                                 videoId="NxZDcqhrdAc", 
                                                 pageToken=response['nextPageToken'], 
                                                 maxResults=2500).execute()
    else:
        break
```

```python
df = pandas.DataFrame(comments)
df.to_excel('results.xlsx', header=['comment', 'author', 'date', 'num_likes'], index=None)
```

##워드클라우드 시각화

```python
!pip install konlpy
```

```python
article_txt = open("/content/drive/MyDrive/galaxyS23(1).txt", "rt", encoding = "UTF-8").read()
```

```python
article_txt
```

```python
from konlpy.tag import Okt
from collections import Counter
```

```python
okt = Okt()
line = []

line = okt.pos(article_txt) 
n_adj = []
for word, tag in line:
    if tag in ["Noun", "Adjective"]: 
        n_adj.append(word)
print(n_adj)
```

```python
#두글자 미만 단어 제거
n_adj2 = []
for i in n_adj :
  if len(i) >= 2:
    n_adj2.append(i)
print(n_adj2)
```

```python
stop_words = "입니다 때문 진짜 정말 같은데 있는데 "
stop_words = set(stop_words.split(" "))
n_adj = [word for word in n_adj2 if not word in stop_words]
```

```python
counts = Counter(n_adj)
tags = counts.most_common(80)
print(counts)
print(tags)
```

```python
from wordcloud import WordCloud
```

```python
font = "/content/drive/MyDrive/malgun.ttf"
word_cloud = WordCloud(font_path = font, background_color = "Black", max_font_size = 100, 
                       colormap = "Blues").generate_from_frequencies(dict(tags))
import matplotlib.pyplot as plt
plt.figure(figsize = (15,12))
plt.imshow(word_cloud)
plt.axis("off")
```
