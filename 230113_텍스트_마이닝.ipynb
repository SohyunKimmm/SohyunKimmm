```python
article_txt = open("/content/drive/MyDrive/article.txt", "rt", encoding = "UTF-8").read()
```

```python
article_txt
```

```python
from konlpy.tag import Okt
from collections import Counter
```

```python
okt = Okt()
line = []

line = okt.pos(article_txt) #article text안에 품사를 붙여주는 함수
n_adj = []
for word, tag in line:
    if tag in ["Noun", "Adjective"]: #명사, 형용사만 뽑음
        n_adj.append(word)
print(n_adj)
```

```python
stop_words = "로 수 더 게 땐 원 것 등 이 명 차 제 후 "
stop_words = set(stop_words.split(" "))
n_adj = [word for word in n_adj if not word in stop_words]
```

```python
counts = Counter(n_adj)
tags = counts.most_common(20)
print(counts)
print(tags)
```

```python
from wordcloud import WordCloud
```

```python
font = "/content/drive/MyDrive/malgun.ttf"
word_cloud = WordCloud(font_path = font, background_color = "black", max_font_size = 300, 
                       colormap = "prism").generate_from_frequencies(dict(tags))
import matplotlib.pyplot as plt
plt.figure(figsize = (10,8))
plt.imshow(word_cloud)
plt.axis("off")
```
