```python
#df 구매유형

import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
 
purchase_data = Counter(df['구매유형']).most_common()
purchase_data = purchase_data[:]
 
data = [x[1] for x in purchase_data] 
purchase = [x[0] for x in purchase_data] 
 

purchase.reverse()
data.reverse()
 

fig =plt.figure(figsize=(8,8))
 
fig.set_facecolor('white') 
colors = sns.color_palette('hls',len(data)) 
plt.yticks(fontsize=15) 
plt.xticks(fontsize=12) 
plt.bar(purchase, data, color= 'yellow',alpha=0.6,edgecolor='k')
plt.title('전체 구매유형',fontsize=20)
plt.show()
```

```python
#vip 구매유형

purchase_data = Counter(vip['구매유형']).most_common()
purchase_data = purchase_data[:]
 
data = [x[1] for x in purchase_data] 
purchase = [x[0] for x in purchase_data] 
 

purchase.reverse()
data.reverse()
 

fig =plt.figure(figsize=(8,8))
 
fig.set_facecolor('white') 
colors = sns.color_palette('hls',len(data)) 
plt.yticks(fontsize=15) 
plt.xticks(fontsize=12) 
plt.bar(purchase, data, color= 'blue', alpha=0.6, edgecolor='k')
plt.title('VIP 구매유형',fontsize=20)
plt.show()
```
